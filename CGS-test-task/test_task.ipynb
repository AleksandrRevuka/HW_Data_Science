**Task 9 (1.5 point).** List top-5 tuples of most correlated **player's characteristics** in the descending order of the absolute [Pearson's coefficient](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient) value.

**Note 1:** Players characteristics are all the columns in `Player_Attributes` table except `[id, player_fifa_api_id, player_api_id, date, preferred_foot, attacking_work_rate, defensive_work_rate]`). <br>
**Note 2:** Exclude duplicated pairs from the list. E.g. ('gk_handling', 'gk_reflexes') and ('gk_reflexes', 'gk_handling') are duplicates, leave just one of them in the resulting list.

```python
player_attributes = pd.read_csv('player_attributes.csv', index_col="Unnamed: 0")
```

```python
player_attributes.head()
```

```python
columns_ignore = ["id", "player_fifa_api_id", "player_api_id", "date", "preferred_foot", "attacking_work_rate", "defensive_work_rate"]
characteristics_columns = [column for column in player_attributes.columns if not column in columns_ignore]

def calculate_correlation(columns_pair: tuple) -> float:
    correlation = player_attributes[list(columns_pair)].corr().iloc[0, 1]
    return correlation

def make_pair_with_correlation(columns: list) -> dict:
    result = {}
    for i, column in enumerate(columns):
        for j in range(i + 1, len(columns)):
            correlation = calculate_correlation((column, columns[j]))
            result[(column, columns[j])] = correlation
    return result
    
pair_with_correlation = make_pair_with_correlation(characteristics_columns)
pair_sorted = dict(sorted(pair_with_correlation.items(), key=lambda item: item[1], reverse=True))

top_correlated_features = list(pair_sorted.keys())[:5]
top_correlated_features
```

**Task 10 (2 points).** Find top-5 most similar players to **Neymar** whose names are given. The similarity is measured as [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance) between vectors of players' characteristics (described in the task above). Put their names in a vector in ascending order by Euclidean distance and sorted by `player_name` if the distance is the same<br>
**Note 1:** There are many records for some players in the `Player_Attributes` table. You need to take the freshest data (characteristics with the most recent `date`). <br>
**Note 2:** Use pure values of the characteristics even if you are aware of such preprocessing technics as normalization. <br>
**Note 3:** Please avoid using any built-in methods for calculating the Euclidean distance between vectors, think about implementing your own.

```python
import numpy as np


neymar_id = player_data[player_data['player_name'] == "Neymar"]["player_api_id"].iloc[0]
neymar_characteristics = player_attributes[player_attributes["player_api_id"] == neymar_id].sort_values(by=["date"], ascending=False).head(1)[characteristics_columns].values

player_attributes_sorted = player_attributes.sort_values(by='date', ascending=False)
player_attributes_grouped = player_attributes_sorted.groupby('player_api_id').first().reset_index()

def euclidean_distance(vec1, vec2):
    return np.sqrt(np.sum((vec1 - vec2) ** 2))

distances = {}

for idx, row in player_attributes_grouped.iterrows():
    player_id = row["player_api_id"]
    player_name = player_data[player_data["player_api_id"] == player_id]["player_name"].iloc[0]
    player_characteristics = row[characteristics_columns].values
    distance = euclidean_distance(neymar_characteristics, player_characteristics)
    distances[player_name] = distance

sorted_distances = sorted(distances.items(), key=lambda x: (x[1], x[0]))

neymar_similarities = sorted_distances[:5]
neymar_similarities
```

**Task 14 (2 points).** Find a team with the highest win rate in the **2015/2016** season. Win rate means won matches / all matches. If there are several teams with the highest win rate return the first by name in lexical order

```python
def winner(row):
    if row["home_team_goal"] > row["away_team_goal"]:
        return row["home_team_api_id"]
    else: 
        return row["away_team_api_id"]
    
columns = ["id", "country_id", "league_id", "season", "stage", "date", "match_api_id", "home_team_api_id", "away_team_api_id", "home_team_goal", "away_team_goal"]

matches_2015_2016 = matches_2015_2016[columns]
matches_2015_2016["team_won_id"] = matches_2015_2016.apply(winner, axis=1)

all_matches_count = matches_2015_2016.shape[0]

team_wins_count = matches_2015_2016["team_won_id"].value_counts()

win_rates = team_wins_count / all_matches_count

team_highest_winrate_2015_2016 = team_data[team_data["team_api_id"] == win_rates.idxmax()]["team_long_name"].iloc[0]
team_highest_winrate_2015_2016
```

**Task 15 (2 points).** Determine the team with the maximum days' gap between matches in **England Premier League 2010/2011 season**. Return number of days in that gap. <br>
**Note**: a *gap* means the number of days between two consecutive matches of the same team.

```python
def count_max_gap_team(matches):
    gap_max = 0
    matches_sorted = matches.sort_values(by="date", ascending=False)
    for i in range(len(matches) - 1):
        gap = matches_sorted.iloc[i]["date"] - matches_sorted.iloc[i + 1]["date"]
        gap_days = gap.days
        if gap_days > gap_max:
            gap_max = gap_days
    return gap_max


epl_league_id = league_data[league_data["league_name"] == "England Premier League"]["league_id"].iloc[0]
epl_matches_2010_2011 = match_data[(match_data["league_id"] == epl_league_id) & (match_data["season"] == "2010/2011")][columns]

comands_id = pd.concat([epl_matches_2010_2011["home_team_api_id"], epl_matches_2010_2011["away_team_api_id"]]).unique()

gap_teams_england_2010_2011 = {}

for team in comands_id:
    home_team_matches = epl_matches_2010_2011[(epl_matches_2010_2011["home_team_api_id"] == team)]
    away_team_matches = epl_matches_2010_2011[(epl_matches_2010_2011["away_team_api_id"] == team)]
    matches_team = pd.concat([home_team_matches, away_team_matches])
    matches_team["date"] = pd.to_datetime(matches_team["date"])
    max_gap = count_max_gap_team(matches_team)
    name_team = team_data[team_data["team_api_id"] == team]["team_long_name"].iloc[0]
    gap_teams_england_2010_2011[name_team] = max_gap

sorted_distances = sorted(gap_teams_england_2010_2011.items(), key=lambda x: (x[1], x[0]), reverse=True)

highest_gap_england_2010_2011 = sorted_distances[0][0]
highest_gap_england_2010_2011
```
