```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
files_idle = [f"idle-{id}.csv" for id in range(1, 1040)]
files_running = [f"running-{id}.csv" for id in range(1, 3409)]
files_stairs = [f"stairs-{id}.csv" for id in range(1, 166)]
files_walking = [f"walking-{id}.csv" for id in range(1, 1851)]

df_idle = pd.DataFrame()
for file in files_idle:
    df = pd.read_csv(file)
    df_idle = pd.concat([df_idle, df], ignore_index=True)

df_running = pd.DataFrame()
for file in files_running:
    df = pd.read_csv(file)
    df_running = pd.concat([df_running, df], ignore_index=True)

df_stairs = pd.DataFrame()
for file in files_stairs:
    df = pd.read_csv(file)
    df_stairs = pd.concat([df_stairs, df], ignore_index=True)

df_walking = pd.DataFrame()
for file in files_walking:
    df = pd.read_csv(file)
    df_walking = pd.concat([df_walking, df], ignore_index=True)

df_idle.to_csv('idle.csv', index=False)
df_running.to_csv('running.csv', index=False)
df_stairs.to_csv('stairs.csv', index=False)
df_walking.to_csv('walking.csv', index=False)
```
